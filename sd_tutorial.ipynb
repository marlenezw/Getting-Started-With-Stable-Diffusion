## Create a save and export GIF function to store your images

```python
def save(image, filename):
    im = Image.fromarray(image)
    im.save(filename)
```

```python
def export_as_gif(images, filename, frames_per_second=10):
    images += images[2:-1][::-1]
    images[0].save(
        filename,
        save_all=True,
        append_images = images[1:],
        duration = 1000 // frames_per_second,
        loop=0
    )

```

## Access the Stable Diffusion model using keras_cv

```python
model = keras_cv.models.StableDiffusion(
    img_width=512,
    img_height=512
)
```

## Generate an image from text

```python
images = model.text_to_image(
    "beautiful Disney Princess, with long black hair, with mountains in the background", 
    batch_size = 3
)

plot(images)
```

## Save the first image in the array to your machine 

```python
save(images[0], 'princess1.jpg')
```
